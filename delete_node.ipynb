```python
import numpy as np
import pandas as pd
```

```python
def distance(B):
        dist_matrix=np.zeros((len(B),len(B)))
        for i in range(len(B)):
                for j in range(len(B)):
                        dist_matrix[i,j]=np.sum(np.abs(B[i]-B[j]))
        return dist_matrix
```

```python
dist_matrix=np.triu(distance(xy))
display(pd.DataFrame(dist_matrix))
```

```python
# 특정 임계값 설정
threshold = 0.0003

# 두 조건을 모두 만족하는 마스크 생성
mask = np.logical_and(dist_matrix < threshold, dist_matrix != 0)

# 조건을 만족하는 값들의 인덱스 찾기
indices = np.where(mask)

# 결과 출력
print("원본 행렬:")
print(dist_matrix)
print("임계값보다 작고 0이 아닌 요소들의 위치와 값:")
edges=[]
# 인덱스와 값을 함께 출력
for i, j in zip(indices[0], indices[1]):
    print(f"위치: ({i}, {j}), 값: {dist_matrix[i, j]}")
    edges.append((i,j))
```

```python
edges
```

```python
import networkx as nx

# 네트워크 생성
G = nx.Graph()

# 모든 지역 0~49를 노드로 추가 (50은 포함되지 않았다고 가정)
for i in range(50):
    G.add_node(i)

# 엣지를 그래프에 추가
for row, col in edges:
    G.add_edge(row, col)

# 연결된 컴포넌트 찾기
connected_components = list(nx.connected_components(G))

# 각 연결된 컴포넌트의 노드 출력
print("각 연결된 컴포넌트의 노드:")
for i, component in enumerate(connected_components, start=1):
    print(f"컴포넌트 {i}: {sorted(component)}")
```
