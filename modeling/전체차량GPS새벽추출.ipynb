```python
df=pd.read_csv("../data/전체차량운행.csv")
```

```python
df=df[["차량번호","보고시각","속도","주소(지번)","주소(POI)","주행거리(m)"]]
df.head(5)
```

```python
df=df.drop_duplicates(subset=['주소(지번)','주행거리(m)'], keep='last')
df
```

```python
df["보고시각"] = pd.to_datetime(df["보고시각"])

# 월, 일, 시, 분 추출
df["월"] = df["보고시각"].dt.month
df["일"] = df["보고시각"].dt.day
df["시"] = df["보고시각"].dt.hour
df["분"] = df["보고시각"].dt.minute
```

```python
df=df.query("속도<=30")
df
```

```python
df=df.query("시<=3 or 시>=17") #17시부터 새벽세시까지
```

```python
df.sort_values(by=["차량번호","보고시각"], ascending=True).to_csv("점검.csv")
```

```python
df['주소(지번)'] = df['주소(지번)'].str.replace(r"\s*\([^)]*\)", "", regex=True)
df
```

```python
df=df.drop_duplicates(subset=['주소(지번)','월','일'], keep='last')
df
```

```python
df.to_csv("전체차량전처리완료.csv")
```
